Creating PDF document

```python
import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Image as RLImage, Spacer
from PIL import Image as PILImage, UnidentifiedImageError
import os

# Load the original CSV file
original_csv = pd.read_csv('sorted_output_file.csv')  # Replace with your file name

# Define the PDF document
pdf_file = 'ESE.pdf'
document = SimpleDocTemplate(pdf_file, pagesize=A4,
                             rightMargin=inch, leftMargin=inch,
                             topMargin=inch, bottomMargin=inch)

# Container for the PDF elements
elements = []

# Table header
table_data = [["Name", "Roll", "Signature"]]

# Number of rows per page
rows_per_page = 16
row_height = 40  # Height of each row in points, same as the image height

# Populate table data
for idx, row in original_csv.iterrows():
    # Append the first and second columns
    row_data = [row[2], row[3]]

    # Load the signature image based on the second column's value
    signature_file = f"{row[3]}.jpg"
    
    if os.path.exists(signature_file):
        try:
            # Check if the image can be opened by PIL
            with open(signature_file, 'rb') as f:
                PILImage.open(f).verify()  # Verify if the image file is valid
            img = RLImage(signature_file, width=100, height=row_height)
            row_data.append(img)
        except (UnidentifiedImageError, IOError) as e:
            print(f"Warning: Could not identify image file '{signature_file}', skipping. Error: {e}")
            row_data.append("")  # Add empty space if image is not valid
    else:
        print(f"Warning: Signature image not found: {signature_file}")
        row_data.append("")

    table_data.append(row_data)

    # Every 16 rows, add the table to the document and start a new one
    if (idx + 1) % rows_per_page == 0:
        table = Table(table_data, colWidths=[2.5*inch, 2.5*inch, 2.5*inch], rowHeights=[row_height]*len(table_data))
        table.setStyle(TableStyle([
            ('BACKGROUND', (0, 0), (-1, 0), colors.white),
            ('TEXTCOLOR', (0, 0), (-1, 0), colors.black),
            ('ALIGN', (0, 0), (-1, -1), 'CENTER'),  # Horizontal alignment
            ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),  # Vertical alignment
            ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
            ('FONTSIZE', (0, 0), (-1, 0), 12),
            ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
            ('BACKGROUND', (0, 1), (-1, -1), colors.white),
            ('GRID', (0, 0), (-1, -1), 1, colors.black),
        ]))
        elements.append(table)
        # elements.append(Spacer(1, 0.5*inch))  # Spacer added between tables
        table_data = [["Name", "Roll", "Signature"]]  # Reset the table with header

# Add the last page of the table if there are remaining rows
if len(table_data) > 1:
    table = Table(table_data, colWidths=[2.5*inch, 2.5*inch, 2.5*inch], rowHeights=[row_height]*len(table_data))
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.white),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.black),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),  # Horizontal alignment
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),  # Vertical alignment
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 12),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, 1), (-1, -1), colors.white),
        ('GRID', (0, 0), (-1, -1), 1, colors.black),
    ]))
    elements.append(table)

# Build the PDF
document.build(elements)

print(f"PDF file '{pdf_file}' created successfully.")
```
